```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels
import matplotlib.pyplot as plt
import math
```

```python
flight = pd.read_csv("flight.csv")
flight.head()
```

# Univariate Analysis

```python
coach_price_data = flight['coach_price']

# Create a histogram using Seaborn
plt.figure(figsize=(8, 6))
sns.histplot(coach_price_data, kde=True, color='skyblue')
plt.title('Histogram of Coach Price')
plt.xlabel('Coach Price')
plt.ylabel('Frequency')
plt.show()

# Calculate the mean and median using Pandas
mean_price = coach_price_data.mean()
median_price = coach_price_data.median()

print("Mean Coach Price:", mean_price)
print("Median Coach Price:", median_price)

plt.clf() # Clear the plot
```

# Bivariate Analysis

```python
coach_prices = flight['coach_price']
first_class_prices = flight['firstclass_price']

# Create a scatter plot using Seaborn
plt.figure(figsize=(8, 6))
sns.scatterplot(x=coach_prices, y=first_class_prices, color='green', alpha=0.02)
plt.title('Relationship between Coach and First-Class Prices')
plt.xlabel('Coach Price')
plt.ylabel('First-Class Price')
plt.show()
```

```python
# Inflight Meals
sns.histplot(flight, x = "coach_price", hue = flight.inflight_meal)
plt.show()
plt.clf()
 
# Inflight Entertainment
sns.histplot(flight, x = "coach_price", hue = flight.inflight_entertainment)
plt.show()
plt.clf()
 
# Inflight WiFi
sns.histplot(flight, x = "coach_price", hue = flight.inflight_wifi)
plt.show()
plt.clf()
```
